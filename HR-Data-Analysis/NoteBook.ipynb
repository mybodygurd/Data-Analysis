```python
df_A.index = 'A' + df_A["employee_office_id"].astype(str)
df_B.index = 'B' + df_B["employee_office_id"].astype(str)
df_HR.index = df_HR["employee_id"].astype(str)
df_A
```

```python
dataAB = pd.concat([df_A, df_B], axis=0, ignore_index=False)
all_data = pd.merge(dataAB, df_HR, how="left", left_index=True, right_index=True, indicator=True)
final_data = all_data[all_data["_merge"] == "both"]
final_data.drop(columns=["employee_id", "employee_office_id", "_merge"], axis=1, inplace=True)
final_data.sort_index(inplace=True)
```

```python
def count_bigger_5(series):
    return (series > 5).sum()

aggregation = {
    "number_project": ["median", count_bigger_5],
    "time_spend_company": ["mean", "median"],
    "Work_accident": ["mean"],
    "last_evaluation": ["mean", "std"]
}

result = final_data.groupby("left").agg(aggregation)
result = result.round(2)
result
# print(result.to_dict())
```

```python
pvt_table = final_data.pivot_table(index="Department",
                       columns=["left", "salary"],
                       values="average_monthly_hours",
                       aggfunc="median"
).round(2)
pvt_table
```

```python
high_col = (0.0, 'high')
medium_col = (0.0, 'medium')
mask1 = pvt_table[high_col] < pvt_table[medium_col]

low_col = (1.0, 'low')
high_col2 = (1.0, 'high')
mask2 = pvt_table[low_col] < pvt_table[high_col2]

# departments = pvt_table.loc[mask1 | mask2].index
# departments
analyze1 = pvt_table.loc[mask1 | mask2]
analyze1
```

```python
pivt_table = final_data.pivot_table(
    index="time_spend_company",
    columns="promotion_last_5years",
    values=["last_evaluation", "satisfaction_level"],
    aggfunc=["min", "max", "mean"]
).round(2)
pivt_table
```

```python
promotion_last_eval = ("mean", "last_evaluation", 1)
no_promotion_last_eval = ("mean", "last_evaluation", 0)
mask = pivt_table[no_promotion_last_eval] > pivt_table[promotion_last_eval]

analyze2 = pivt_table.loc[mask]
analyze2
```
