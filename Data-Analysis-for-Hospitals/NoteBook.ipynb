```python
import pandas as pd
pd.set_option('display.max_columns', 15)
```

```python
general_df = pd.read_csv("artifacts/general.csv")
prenatal_df = pd.read_csv("artifacts/prenatal.csv")
sports_df = pd.read_csv("artifacts/sports.csv")
general_df.head(20)
```

```python
prenatal_df.head(20)
```

```python
sports_df.head(20)
```

```python
suitable_column_names = general_df.columns
prenatal_df.columns = suitable_column_names
sports_df.columns = suitable_column_names

dataframe = pd.concat([general_df, prenatal_df, sports_df], axis=0, ignore_index=True)

dataframe = dataframe.drop(columns=["Unnamed: 0"], axis=1)

dataframe.sample(n=20, random_state=30)
```

```python
print(f"Data Shape: {dataframe.shape}")
```

```python
dataframe.isnull().sum()
```

```python
general_df[general_df["gender"].isna()]
```

```python
sports_df[sports_df["gender"].isna()]
```

```python
dataframe = dataframe.dropna(axis="rows", how="all")
dataframe.isnull().sum()
```

```python
dataframe["gender"].fillna("f", inplace=True)

dataframe["gender"].replace({
    "female": "f",
    "woman": "f",
    "male": "m",
    "man": "m"
}, inplace=True)
dataframe.head(10)
```

```python
columns = ["bmi", "diagnosis", "blood_test", "ecg", "ultrasound", "mri", "xray", "children", "months"]
dataframe[columns] = dataframe[columns].fillna(0)

print(f"Data Shape: {dataframe.shape}")
```

```python
dataframe.isnull().sum()
```

```python
dataframe.sample(n=20, random_state=30)
```

```python
dataframe["hospital"].value_counts().idxmax()
```

```python
dataframe.columns
```

```python
general_hosp = dataframe.query("hospital == 'general'")
stomach_cases = general_hosp.query("diagnosis == 'stomach'").shape[0]
share = round((stomach_cases / len(general_hosp)) * 100, 3)
share
```

```python
sports_hosp = dataframe[dataframe["hospital"] == "sports"]
share = round(
    (sports_hosp["diagnosis"].value_counts(normalize=True).get("dislocation", 0)) * 100, 3
)
share
```

```python
sports_hosp = dataframe[dataframe["hospital"] == "sports"]
unknown_cases = (sports_hosp["diagnosis"] == "dislocation").sum()
share = round((unknown_cases / len(sports_hosp)) * 100, 3)
share
```

```python
general_hosp_median_age = general_hosp["age"].median()
sports_hosp_median_age = sports_hosp["age"].median()
age_diff = general_hosp_median_age - sports_hosp_median_age
age_diff
```

```python
pivt = dataframe.pivot_table(
    index="hospital",
    columns="blood_test",
    aggfunc="count"
)
blood_test_counts = pivt["age"]["t"]
blood_test_counts.idxmax()
```
